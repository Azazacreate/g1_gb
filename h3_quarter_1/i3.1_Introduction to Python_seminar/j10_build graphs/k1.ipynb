```python
import pandas as pd
import seaborn as sea


df_2 = pd.read_csv("penguins-2.csv")

# добавить поле:новое в-DataFrame 
df_2.loc[df_2["bill_length_mm"] < 35, "height_group"] = "low"
df_2.loc[(df_2["bill_length_mm"] >= 35) & (df_2["bill_length_mm"] <= 42), "height_group"] = "middle"
df_2.loc[df_2["bill_length_mm"] > 42, "height_group"] = "high"

# надо-добавить изменения в-файл.
df_2.to_csv("penguins-2.csv", index="False")

sea.displot(data=df_2, x="bill_length_mm", hue="height_group", col="sex")
```
